```python
center_path = r"C:\\Users\\svkar\\Desktop\\python workspace\\fulfilment_center_info.csv"
df_center = pd.read_csv(center_path)
df_center.head(50)
```

```python
train_path = r"C:\\Users\\svkar\\Desktop\\python workspace\\train.csv"
df_food = pd.read_csv(train_path)
df_food
```

```python
df3=pd.merge(df_food,df_center,on='center_id')
```

```python
df3
```

```python
df=pd.merge(df_meal,df3,on='meal_id')
```

```python
df
```

```python
table = pd.pivot_table(data=df,index='category',values='num_orders',aggfunc=np.sum)

type(table)
table
```

```python
table.plot(kind='bar')
plt.xlabel('Food items')
plt.ylabel('Quantity sold')
plt.xticks(rotation=70)
#we use xticks to rotate the xaxis label by 70 
```

```python
df['revenue']=df["checkout_price"]*df["num_orders"]
df.revenue
df
```

```python
month1=df.week/4

month1=month1.astype(int)
month1
```

```python
month=month1.astype(float)
month
```

```python
df['month']=month
df
```

```python
table1 = pd.pivot_table(data=df,index='month',values='revenue',aggfunc=np.sum)

type(table1)
table1
```

```python
table2 = pd.pivot_table(data=df,index='week',values='revenue',aggfunc=np.sum)

type(table2)
table2
```

```python
fig = plt.figure(figsize=(8,4))
ax1 = fig.add_subplot(121)
table1.plot(kind='line')
plt.title('monthly revenue')

ax1 = fig.add_subplot(122)
table2.plot(kind = 'line')
plt.title('weekly revenue')
```
